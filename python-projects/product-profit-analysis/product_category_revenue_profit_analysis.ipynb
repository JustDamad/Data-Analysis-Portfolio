```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl

df = pd.read_excel(r"C:\Users\JustDamad\Documents\Top_Product_Sales.xlsx")
df 
```

```python
df.info()
df.isnull().sum() 
```

```python
# check if any missing values

df = df.dropna() 
```

```python
# Check  Revenue and Profit Formulas

df["REVENUE"] = df["UNITS_SOLD"] * df["UNIT_PRICE"]
df["PROFIT"] = df["REVENUE"] - df["TOTAL_COST"] 
```

```python
# check revenue and profit per product

rev = df[["PRODUCT", "REVENUE"]].sort_values(by = "REVENUE", ascending = False)
profit = df[["PRODUCT", "PROFIT"]].sort_values(by = "PROFIT", ascending = False)
```

```python
# Analyze & Group

top5_rev = rev.head(5)
top5_profit = profit.head(5)
top5_rev, top5_profit 
```

```python
# Visualize revenue chat

plt.figure(figsize = (10,6))
plt.bar(top5_rev["PRODUCT"], top5_rev["REVENUE"])
plt.title("Top 5 PRODUCTS by REVENUE")
plt.xlabel("PRODUCT") 
plt.ylabel("REVENUE")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show() 
```

```python
# Visualize Profit Chart

plt.figure(figsize = (10,6))
plt.bar(top5_profit["PRODUCT"], top5_profit["PROFIT"])
plt.title("TOP 5 PRODUCTS by PROFIT")
plt.xlabel("PRODUCT")
plt.ylabel("PROFIT")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show() 
        
```

# INSIGHTS

* Smartphone generated the highest revenue(60,000).
* Smartphone also produced the highest profit(24,000).
* Laptop followed closely in both REVENUE and PROFIT.
* Accessories (Keyboard, Router, Power Bank) contribute steady but smaller profits
* Products with HIGHER UNIT PRICES tend to dominate profit ranking
