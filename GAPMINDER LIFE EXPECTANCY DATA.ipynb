# GAPMINDER LIFE EXPECTANCY ANALYSIS (1952-2007)

### Objective
To analyze change in Life expectancy across Nigeria, India and Ghana over multiple decades.

### Tools Used
- Python
- Pandas
- Seaborn
- Matplotlib

### Key Steps 
- Filtered dataset by selected countries
- Created a line chart to show life expectancy over time

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
```

```python
tips = sns.load_dataset("tips") 
```

```python
tips.head() 
```

```python
sns.get_dataset_names() 
```

```python
sns.color_palette("pastel")
sns.scatterplot(x = "total_bill", y = "tip", data = tips) 
```

```python
plt.figure(figsize = (6,5))
sns.set_theme(style = "darkgrid", palette = "Set2", font_scale = 1.2)
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "sex", 
                style = "time", size = "size") 
```

```python
sns.relplot(x = "total_bill", y = "tip" , data = tips, kind = "scatter", 
            hue = "sex", style = "time") 
```

```python
import plotly.express as px 
```

```python
pd = px.data.gapminder() 
```

```python
pd.head() 
```

```python
pd_india = pd[pd["country"] == "India"] 
```

```python
pd_india.head() 
```

```python
plt.figure(figsize = (6,5))
sns.lineplot(x = "year", y = "lifeExp", data = pd, ci = False)
plt.show() 
```

```python
pd_compare = pd[pd["country"].isin(["Nigeria", "India", "Ghana"])] 
```

```python
pd_compare.head() 
```

```python
sns.relplot(x = "year", y = "lifeExp", data = pd_compare, hue = "country", 
            kind = "line") 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (10,6)) 
sns.lineplot(x = "year", y = "lifeExp", data=pd_compare, hue = "country")
plt.title("Life Expectancy Trends: Nigeria vs India vs Ghana")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.show() 
```

### INSIGHTS
- Overtime, life expectancy has been steadily increasing in all three countries
- India and Ghana show slightly faster improvement compared to Nigeria.
- India shows the most significant improvement over time.
- Nigeria maintains the lowest life expectancy throughout the years,
  highlighting potential gaps in healthcare and living conditions.

### CONCLUSION
Healthcare progress is improving, but Nigeria needs more investment to close the gap.
