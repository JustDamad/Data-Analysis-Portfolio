```python
import sqlite3
import pandas as pd
```

```python
data = {"Transaction_ID": [1,2,3,4,5],
        "Customer_ID": ["CUST001", "CUST002", "CUST001", "CUST003", "CUST002"],
        "Country": ["Nigeria", "Ghana", "Nigeria", "South Africa", "Nigeria"],
        "Transaction_Type": ["Transfer", "Bills", "Airtime", "Transfer", "Bills"],
        "Channel": ["USSD", "Mobile App", "POS", "USSD", "Agent"],
        "Amount": [25000, 15000, 5000, 30000, 12000],
        "Fee": [50, 30, 10, 60, 20],
        "Date": ["2024-01-10", "2024-01-11", "2024-01-13", "2024-02-01", "2024-02-03"],
        "Status": ["Success", "Success", "Failed", "Success", "Success"]
} 

df = pd.DataFrame(data) 
```

```python
# save the data to SQLite Database

db_path = "Fintech_Transactions.db"
conn = sqlite3.connect(db_path)
df.to_sql("transactions", conn, if_exists = "replace", index = False)
conn.close() 

print("Database created successfully:", db_path) 
```
