## Use a pivot table (reshape the summerrized data) to create a heatmap ## 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
gapminder = pd.read_csv(url) 

pd_pivot = gapminder.pivot(index="country", columns="year", values="lifeExp")

pd_pivot 
```

```python
plt.figure(figsize = (12,10))
sns.heatmap(pd_pivot, cmap="viridis")  # <-- actual plot
plt.title("Life Expectancy by Country and Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Country")

plt.show() 
```

```python
### show heatmap for only africa countries ### 

pd_africa = gapminder[gapminder["continent"] == "Africa"]

afri_pivot = pd_africa.pivot(index ="country", columns = "year", values = "lifeExp")
plt.figure(figsize=(12, 10))
sns.heatmap(afri_pivot, cmap="viridis")
plt.title("Life Expectancy in Africa (19522007)")
plt.xlabel("Year")
plt.ylabel("Country")
plt.show() 
```

```python
sns.heatmap(afri_pivot, cmap="viridis", linewidths=0.3)  
```







## INSIGHTS ##

*  The heatmap shows long-term progress in public health and living conditions.  
*  Not all of Africa advanced at the same pace(highlights regional inequality).
*  The heatmap visually encodes the historical and social challenges that shaped African health outcomes. #

## CONCLUSION ##

*  The heatmap shows that life expectancy in African countries has steadily increased from 1952 to 2007, though progress has been uneven across regions, with some nations improving faster while others were slowed by conflict, disease, or economic challenges
